# Key Understanding About Python

**Python Code Execution Steps**

1. **Lexical Analysis**  
   The Python interpreter breaks the source code into smaller parts called **tokens**, such as keywords, identifiers, numbers, symbols, and punctuation.

2. **Syntax Parsing**  
   Python checks whether the structure of the code is correct. If the syntax rules are violated, a **SyntaxError** is raised.

3. **Bytecode Generation**  
   After successful syntax checking, the code is converted into a simpler set of instructions called **bytecode**.

4. **Execution by PVM**  
   The **Python Virtual Machine (PVM)** executes the bytecode and performs operations like calculations, value assignments, and function calls.

5. **Error Handling and Output**  
   During execution, Python handles runtime errors if any occur and displays the output such as printed messages or returned values.


**Interpretor**
1. It tends to translate and execute code line by line.
2. Execution: Line by line
3. Speed: Slower
4. Debugging: Easier
5. Example: Python, JavaScript, Ruby and PHP

**Compiler**
1. It tends to translate the entire code into machine code before program runs
2. Execution: Entire program at once
3. Speed: Faster
4. Debugging: Complex
5. Example: C, C++, Java

# Type Casting

**Type Casting** : It used to convert the data from one type to another

- All **Str** type can't be casted into numerical type

- Types of casting: Implicit type casting, Explicit type casting

**Implicit Type Casting** :  It gets perfom automatically by the python interpreter, this usually gets performing operation between different data types

**Explicit Type Casting** : It known as type conversion, is performed mannually by the programmer using built-in functions, ensure the desired type conversions


```python
a = 1
print(type(a))

b = "1"
#print(a + b) # This will raise a TypeError because you cannot add an integer and a string directly

c = int(b)
print(a + c) # This will correctly add the integer values of a and c

print(a + int(b)) #  This will raise a TypeError because you cannot add an integer and a string directly


mynum = 26
mynum2 = str(mynum) # converting integer to string
print(mynum2) # printing the string representation of the integer


# Implicit  type casting 
var1 = 10 # integer
var2 = 5.5 # float
var3 = var1 + var2  # var1 is implicitly converted to float
print(var3)  # Output: 15.5
print(type(var3))  # Output: <class 'float'>

# Explicit type casting
var4 = 10.7 # float 
var5 = int(var4)  # explicitly converting float to integer
```

```python
# Create a Calculator Function

num1 = int(input("Enter first number: "))  # taking user input for first number
num2 = int(input("Enter second number: "))  # taking user input for second number
operation = input("Enter operation (add, subtract, multiply, divide, average): ")  # taking user input for operation


def calculator(num1, num2, operation):
    if operation == "add":
        return num1 + num2
    elif operation == "subtract":
        return num1 - num2
    elif operation == "multiply":
        return num1 * num2
    elif operation == "divide":
        if num2 != 0:
            return num1 / num2
        else:
            return "Error: Division by zero is not allowed."
    elif operation == "average":
        return (num1 + num2) / 2
    else:
        return "Error: Invalid operation."
calculator_result = calculator(num1, num2, operation)  # calling the calculator function to add num1 and num2
print("Calculator Result:", calculator_result)  # printing the result of the calculator function
```

## Function Arguement()

1. Argument are the value that are passed into a function when it's called.
2. If a function has 2 parameters, you must  provide 2 arhuments when calling it.

**Types of Function Argument**
1. Required Argument (Single/Mutiple): function defined using one parameter (variables)
2. Default Argument: function defined using one parameter & default value
3. Keyword Argument (Named): function defined using two parameters
4. Arbitary Arguments (Types *args (wish to pass multiple values: *positional Argument* store as tuple) **kwargs (Multiple keyword Argument *keyword Argument* store as dictionary)




```python
#Required Argument Example

def greeting(name): #name is parameter
    print("Hello,", name + "! Welcome to Python scripting practice.")
greeting("Rahul")  # Rahul as argument
```

```python
greeting() # Required an argument to run code
```

```python
# Required Argument Example

def intro(course_name , instructor_name):
    print("Welcome to the", course_name, "course conducted by", instructor_name)

intro("Pyhton for Machine Learning!!", "Rahul G")
```

```python
# Default Argument Example

def greeting(name = "World!!"):
    print("Hello my friend, Welcome to my Technical", name)

greeting()
```

```python
# Keyword Argument Example

def divide(a,b): # a, b are parameters
    return a/b
result = divide(a=10, b=20) #with keyword arguments & positional arguments
print(result)
```

```python
# Arbitrary Argument Example : used when we dont know how many arguments will be passed

def add_number(*args): #args is parameter
    return sum(args)

# Any number of arguments can be passed
result = add_number(1, 2, 3, 4)
print(result) 
```

```python
# Arbitrary Argument Example
def greetings(*names):
    for name in names:
        print(f"Hello, {name}! Welcome to Python scripting practice.")

greetings("Alice", "Bob", "Charlie")
```

```python
# Arbitrary Argument Example
def cards(*cards):
    for card in cards:
        print(f"You have a {card} card.")

cards("Visa", "MasterCard", "Amex")
```

```python
# Arbitrary Argument Example **kwargs

def student_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

        
student_info(name="Rahul", age=21, city="Chicago")
```

```python
# Arbitrary Argument Example

def passenger_info(**details):
    for key, value in details.items():
        print(f"{key}: {value}")

passenger_info(name="Alice", age=30, destination="New York")
```
