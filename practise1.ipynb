```python
#putting arithmetic operations with variables
b = 10

c = 20

print(b+c) # printing the sum of two variables b and c
print("The sum of", b, "and", c, "is", b + c) # printing a descriptive message along with the sum of b and c

print(c/b) # printing the result of division of c by b
print("The division of", c, "by", b, "gives", c / b) # printing a descriptive message along with the result of division

print(c*b) # printing the result of multiplication of c and b
print("The multiplication of", c, "and", b, "results in", c * b) # printing a descriptive message along with the result of multiplication

print(c-b) # printing the result of subtraction of b from c
print("The subtraction of", b, "from", c, "yields", c - b) # printing a descriptive message along with the result of subtraction    
```
